```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### 1 layer transformer

```python
n_embed = [8, 12, 16, 24, 32, 64, 128, 256]
n_heads = [1, 2, 4, 6, 8, 12, 16, 20, 24]
```

```python
running_avg = np.array([[0, 0, 0, 7, 98.6, 99.78, 100, 100],
                        [3.64, 26.76, 99.9, 100, 100, 100, 100, 80.66],
                        [99.76, 99.76, 99.98, 100, 100, 100, 100, 6.98],
                        [99.88, 100, 100, 100, 100, 100, 100, 14.86],
                        [100, 99.98, 100, 100, 100, 100, 99.98, 0.02],
                        [99.98, 100, 100, 100, 100, 100, 40, 0.06],
                        [100, 100, 100, 100, 100, 60, 0, 0],
                       [100, 100, 100, 100, 100, 100, 80, 0],
                       [100, 100, 100, 100, 100, 90, 65, 0]])
```

```python
sns.heatmap(running_avg, xticklabels=n_embed, yticklabels=n_heads)
plt.xlabel('Per-head embed dim')
plt.ylabel('Nb of heads')
plt.title('Phase transition graph for 1 layer')
plt.show()
```
